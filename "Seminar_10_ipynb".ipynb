```python
import pandas as pd
import numpy as np
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(7)
```

```python
data['human']=np.where(data['whoAmI']=='human',1, 0)
data['robot']=np.where(data['whoAmI']=='robot',1, 0)
data.head(7)
```

```python
data['human']=np.where(data['whoAmI']=='human', True, False)
data['robot']=np.where(data['whoAmI']=='robot', True, False)
data.head(7)
```
